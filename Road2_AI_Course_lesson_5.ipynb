now, please write code that creates a copy of this dataset, but does the following things to each image:

resize to 150 by 150 pixels

turn to gray scale

prints a counter of the amount of images proccessed so far

the new dataset is to be saved in this folder: /content/drive/MyDrive/Colab Notebooks/lesson 5/dataset

```python
import os
import cv2
from tqdm import tqdm

dataset_path = '/content/drive/MyDrive/Colab Notebooks/lesson 5/cats_and_dogs_filtered'
new_dataset_path = '/content/drive/MyDrive/Colab Notebooks/lesson 5/dataset'

# Function to process and save images
def process_images(input_path, output_path):
    if not os.path.exists(output_path):
        os.makedirs(output_path)

    for category in os.listdir(input_path):
        category_path = os.path.join(input_path, category)
        output_category_path = os.path.join(output_path, category)

        if not os.path.exists(output_category_path):
            os.makedirs(output_category_path)

        for image_name in tqdm(os.listdir(category_path), desc=f"Processing {category} images"):
            image_path = os.path.join(category_path, image_name)
            output_image_path = os.path.join(output_category_path, image_name)

            # Read the image
            img = cv2.imread(image_path)

            # Resize to 150x150 pixels
            img_resized = cv2.resize(img, (150, 150))

            # Convert to grayscale
            img_gray = cv2.cvtColor(img_resized, cv2.COLOR_BGR2GRAY)

            # Save the processed image
            cv2.imwrite(output_image_path, img_gray)

    print(f"\nProcessing complete. Images saved in: {output_path}")

# Process and save images in the training set
process_images(os.path.join(dataset_path, 'train'), os.path.join(new_dataset_path, 'train'))

# Process and save images in the testing set
process_images(os.path.join(dataset_path, 'validation'), os.path.join(new_dataset_path, 'validation'))
```

please create a convolution neural network using keras. the input of the network should be an image from the new dataset in this folder: /content/drive/MyDrive/Colab Notebooks/lesson 5/dataset. the following layers of the network are:

a convolution layer after padding the image, using 32 kernels in a size of 3 by 3, then a relu function, and then max pooling using a 2 by 2 kernel with a stride of 2

another convolution layer after padding the image, using 32 kernels in a size of 3 by 3, then a relu function, and then max pooling using a 2 by 2 kernel with a stride of 2

another convolution layer after padding the image, using 32 kernels in a size of 3 by 3, then a relu function, and then max pooling using a 2 by 2 kernel with a stride of 2

a softmax layer that chooses the most probable option as the output

after creating the network, please print its architecture

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Create the CNN model
model = Sequential()

# First convolutional layer
model.add(Conv2D(32, (3, 3), padding='same', activation='relu', input_shape=(150, 150, 1)))
model.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))

# Second convolutional layer
model.add(Conv2D(32, (3, 3), padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))

# Third convolutional layer
model.add(Conv2D(32, (3, 3), padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))

# Flatten layer to transition from convolutional to fully connected layers
model.add(Flatten())

# Fully connected layer with softmax activation for classification
model.add(Dense(2, activation='softmax'))

# Print the model architecture
model.summary()
```

now, please train the network over 20 epoch in batches of 50, using adam's optimizer, using only the training section of both cats and dogs of the new dataset you created in this folder: /content/drive/MyDrive/Colab Notebooks/lesson 5/dataset. in the end, print the network's learning curve

```python
from keras.preprocessing.image import ImageDataGenerator

# Define the paths
train_data_dir = '/content/drive/MyDrive/Colab Notebooks/lesson 5/dataset/train'
batch_size = 50
epochs = 20

# Data augmentation to increase the diversity of training data
train_datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)

train_generator = train_datagen.flow_from_directory(
    train_data_dir,
    target_size=(150, 150),
    batch_size=batch_size,
    color_mode='grayscale',
    class_mode='categorical'
)

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // batch_size,
    epochs=epochs
)

# Plot the learning curve
import matplotlib.pyplot as plt

plt.plot(history.history['accuracy'])
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.show()
```

please test the network using the enitre testing section in the dataset you created before in this folder: /content/drive/MyDrive/Colab Notebooks/lesson 5/dataset, and print a confusion matrix of the network's results

```python
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report
from keras.preprocessing.image import ImageDataGenerator

# Define the path for the testing data
test_data_dir = '/content/drive/MyDrive/Colab Notebooks/lesson 5/dataset/validation'

# Data generator for testing data
test_datagen = ImageDataGenerator(rescale=1./255)

test_generator = test_datagen.flow_from_directory(
    test_data_dir,
    target_size=(150, 150),
    batch_size=50,
    color_mode='grayscale',
    class_mode='categorical',
    shuffle=False
)

# Evaluate the model on the testing set
eval_result = model.evaluate(test_generator)

# Make predictions on the testing set
predictions = model.predict(test_generator)
predicted_labels = np.argmax(predictions, axis=1)

# True labels
true_labels = test_generator.classes

# Confusion matrix
conf_matrix = confusion_matrix(true_labels, predicted_labels)
print("Confusion Matrix:")
print(conf_matrix)

# Classification report
print("\nClassification Report:")
print(classification_report(true_labels, predicted_labels))
```
